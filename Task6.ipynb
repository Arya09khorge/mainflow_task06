```python
df['type'].value_counts().plot(kind='bar',color=['skyblue','orange'])
plt.title('Distribution of movies and TV shows')
plt.xlabel('type')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
#movies has more distribution compared to TV shows
```

```python
#Question 2: Distribution of Ratings
rating_count = df['rating'].value_counts()
plt.figure(figsize=(5,5))
sns.barplot(x=rating_count.index, y=rating_count.values, palette='viridis')
plt.title('Distribution of Ratings')
plt.xlabel('rating')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Question 3: Top 10 Directors
directors = df['director'].astype(str).str.split(',').explode().str.strip()
directors = directors[directors != 'Unavailable']
director_counts = directors.value_counts()
top_10_directors = director_counts.head(10)
plt.figure(figsize=(20,12))
sns.barplot(x=top_10_directors.values,y=top_10_directors.index,legend=False,palette='viridis')
plt.title('Top 10 directors with the most content')
plt.xlabel('Number of Contents')
plt.ylabel('Director')
plt.show()
```

```python
#Question 4: Top 10 TV Shows with the most seasons
tv_shows = df[df['type']=='TV Show']
tv_shows['num_seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(int)
tv_shows_sorted = tv_shows.sort_values(by='num_seasons',ascending=False)
top_10_tv_shows = tv_shows_sorted.head(10)
plt.figure(figsize=(20,12))
sns.barplot(x='num_seasons',y='title',data=top_10_tv_shows,palette='viridis')
plt.title('Top 10 TV shows with the most seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('TV Show')
plt.show()
```

```python
#Question 5: Top 10 Actors/Actresses
cast=df['cast'].str.split(',').explode().str.strip()
cast=cast[cast!='Unavailable']
top_10_actors=cast.value_counts().head(10)
plt.figure(figsize=(20,12))
sns.barplot(x=top_10_actors.values,y=top_10_actors.index,palette='viridis')
plt.title('Top Actor/Actresses by Frequency')
plt.xlabel('count')
plt.ylabel('Actor/Actress')
plt.show()
```

```python
#Question 6: Distribution of release years for movies and shows
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='release_year',hue='type',kde=True,bins=20)
plt.title('Distribution of release years for movies and TV shows')
plt.xlabel('release year')
plt.ylabel('count')
plt.legend(title='type')
plt.show()
```
